# If the file is with simplest data

# PyMudPDFLoader
# This type of loader is efficience and reliable option for loading pdf, espacially when performance and advance features (password) are importatn

```python
from langchain_community.document_loaders import PyMuPDFLoader
loader = PyMuPDFLoader("example_data/layout-parser-paper.pdf")
data = loader.load()
```

# PyPDFium2Loader
- Advanced features like form field extraction or password-protected document handling are required, PyMuPDFLoader might be a better option.
- In general, PyPDFium2Loader is a good all-around choice for loading PDF documents, especially when performance and simplicity are important

```python
from langchain_community.document_loaders import PyPDFium2Loader
loader = PyPDFium2Loader("example_data/layout-parser-paper.pdf")
data = loader.load()
```

# If the pdf file contains much of mathematical equations then use this
